```python
# From https://www.tensorflow.org/get_started/get_started
from functools import partial as bind
import tensorflow as tf
assert(tf.__version__.find('2') == 0)

# training data
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

# Model parameters
W = tf.Variable([0.3], tf.float32)
b = tf.Variable([-0.3], tf.float32)

# Our hypothesis XW+b
hypothesis = lambda X: X * W + b

# cost/loss function
@tf.function
def cost(X, Y):
    return tf.reduce_sum(tf.square(hypothesis(X) - Y))

# Init Gradient Descent optimizer
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

# training
for step in range(1000):
    # Minimize
    optimizer.minimize(bind(cost, x_train, y_train), var_list=[W, b])

    # evaluate training accuracy
    tf.print("W: ", W, "b: ", b, "cost: ", bind(cost, x_train, y_train)())

"""
W: [-0.9999969] b: [0.9999908] cost: 5.699973826267524e-11
"""
```
